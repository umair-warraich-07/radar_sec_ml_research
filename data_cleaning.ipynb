```python
import pandas as pd
```

```python
df_dataset = pd.read_csv("processed_dataset_friday.csv",low_memory=False)
```

```python
df_dataset
```

```python
df_dataset.columns
```

```python
df_dataset["Label"].unique()
```

### 2- Removing row with erroneous 'Label' label

```python
# Remove erroneous 'Label' row
df_dataset.drop(df_dataset.loc[df_dataset["Label"] == "Label"].index, inplace=True)
```

```python
df_dataset
```

```python
df_dataset["Label"].unique()
```

### 3- Creating "Malicious" label from the "Bot", "DoS attacks-SlowHTTPTest", and "DoS attacks-Hulk" labels

```python
df_dataset.replace(to_replace=['Infilteration'], value="Malicious", inplace=True)
```

```python
df_dataset
```

```python
df_dataset["Label"].unique()
```

```python
print(df_dataset["Label"].value_counts()[['Benign']].sum())
print(df_dataset["Label"].value_counts()[['Malicious']].sum())
```

### 4- Checking "Protocol" values, and creating dummies based on the assumption that the Protocol values are categorical

```python
df_dataset["Protocol"].unique()
```

```python
# Forcing them to be strings
df_dataset = df_dataset.astype({"Protocol": str})
```

```python
df_dataset["Protocol"].unique()
```

```python
df_dataset = pd.get_dummies(df_dataset, columns=['Protocol'], drop_first=True)
```

```python
df_dataset
```

```python
# making Label column the last column again
df_dataset.insert(len(df_dataset.columns)-1, 'Label', df_dataset.pop('Label'))
```

```python
df_dataset
```

### 5- Dropping some columns that may be irrelevant (would require re-analysis)

```python
columns_to_drop = [
    'Flow Byts/s',  # This field had np.inf values during training, as such was removed
    'Flow Pkts/s'  # This field had np.inf values during training, as such was removed
]
```

```python
df_dataset.drop(columns=columns_to_drop, inplace=True)
```

```python
df_dataset
```

### 6- Dropping missing or duplicate values

```python
df_dataset.dropna(inplace=True)
```

```python
df_dataset
```

```python
df_dataset.drop_duplicates(inplace=True)
```

```python
df_dataset
```

### 7- Checking distribution of labels after dropping missing or duplicate values

* Malicious: 21.260%
* Benign: 78.739%

```python
print(df_dataset["Label"].value_counts()[['Benign']].sum())
print(df_dataset["Label"].value_counts()[['Malicious']].sum())
```

### 8- Replacing Benign labels with zeroes and Malicious labels with ones 

```python
df_dataset.replace(to_replace="Benign", value=0, inplace=True)
df_dataset.replace(to_replace="Malicious", value=1, inplace=True)
```

```python
df_dataset
```

```python
print(df_dataset["Label"].value_counts()[[0]].sum())
print(df_dataset["Label"].value_counts()[[1]].sum())
```
